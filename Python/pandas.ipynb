```python
# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
vis1=sns.distplot(df['InternetUsers'])
plt.show()
```

```python
vis2=sns.displot(df["InternetUsers"])
```

```python
vis3=sns.distplot(df["InternetUsers"],bins=15)
plt.show(vis3)
```

```python
vis4=sns.boxplot(data=df,x="IncomeGroup",y="BirthRate")
plt.show(vis4)
```

```python
## outlier -- stats outlier is the data point is very far from other observation(ANOMALY DETECTION) 
```

```python
vis5=sns.lmplot(data=df,x="InternetUsers",y="BirthRate")
plt.show(vis5)
```

```python
vis5=sns.lmplot(data=df,x="InternetUsers",y="BirthRate",fit_reg=False)
plt.show(vis5)
```

```python
vis5=sns.lmplot(data=df,x="InternetUsers",y="BirthRate",fit_reg=True)
plt.show(vis5)
```

```python
vis8=sns.lmplot(data=df,x="InternetUsers",y="BirthRate",fit_reg=False,hue="IncomeGroup") #hue - parameter for colour
plt.show(vis8)
```

```python
vis9=sns.lmplot(data=df,x="InternetUsers",y="BirthRate",fit_reg=True,hue="IncomeGroup") #hue - parameter for colour
plt.show(vis9)
```
